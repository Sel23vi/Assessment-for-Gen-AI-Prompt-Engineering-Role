```python
# Install necessary libraries
# !pip install pandas numpy matplotlib seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset (replace 'titanic.csv' with your file path)
df = pd.read_csv('titanic.csv')

# Display the first few rows
df.head()
```

```python
## Summary statistics

# Basic descriptive statistics
df.describe()

# Information about data types and missing values
df.info()

# Check for missing values
df.isnull().sum()
```

```python
# Specific Insights:

# 1. Distribution of Age:
# Basic statistics for the 'Age' column
print("Mean Age:", df['Age'].mean())
print("Median Age:", df['Age'].median())
print("Age Standard Deviation:", df['Age'].std())
```

```python
# 2. Survival Rate by Gender:
# Group by gender and calculate survival rate
survival_rate_by_gender = df.groupby('Sex')['Survived'].mean()
print(survival_rate_by_gender)
```

# 3. Data Visualization

```python
# Plot 1: Age Distribution

# Set up Seaborn style
sns.set(style='whitegrid')

# Plot age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Passengers', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

```python
# Plot 2: Survival Rate by Passenger Class (Pclass)

# Plot survival rate by Pclass
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Passenger Class', fontsize=16)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.show()
```

```python
# Plot 3: Gender-Based Survival Rate

# Plot survival rate by gender
plt.figure(figsize=(8, 5))
sns.barplot(x='Sex', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.show()
```

```python
# Combining the distribution of a numerical variable with a hue of categorical variable

sns.displot(data = df, x = 'Fare', hue = 'Survived', kind = 'kde')
plt.show()
```

```python
# Plot 4: Heatmap for Correlation

# Compute correlation matrix
corr_matrix = df.drop(['PassengerId', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1).corr()

# Plot heatmap for correlation
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix', fontsize=16)
plt.show()
```
